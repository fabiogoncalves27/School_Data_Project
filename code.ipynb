```python
# Which 10 states have the most number of schools that opened projects to gather donations ? 
s = schools["School State"].value_counts().sort_values(ascending = False).head(10)
print(s)

# Visualize with a bar plot
s.plot.bar()
plt.xlabel("States")
plt.ylabel("Number of schools")
plt.title("Number of Schools involved in Projects by State")
plt.tight_layout()
plt.margins(0.05)
plt.show()
```

```python
# What are the top 10 states in which schools gathered most amount of AVERAGE donations for their projects ? 
data = pd.merge(projects, donations, how="inner", on="Project ID")
data2 = pd.merge(data, schools, how="inner", on="School ID")

s1 = data2.groupby("School State")["Donation Amount"].mean().round(2).sort_values(ascending = False).head(10)
print(s1)
```

```python
# Visualize with a bar plot visualize states that have more average donations per project than others.
s1.plot.bar()
plt.xlabel("States")
plt.ylabel("Average donations per project")
plt.title("Top 10 average donations per project by State")
plt.tight_layout()
plt.margins(0.05)
plt.show()
```

```python
# Investigate the maximum, minimum, mean, median, 25th and 75th percentiles of "Donation Amount" column. 
# What is the average donation amount across all projects ? 
# What are the minimum and maximums ?

# Using descriptive statistics
print(data2["Donation Amount"].describe())

# Get maximun donation amount
print("Maximum donation amount is :",data2["Donation Amount"].max())

# Get minimum donation amount
print("Minimum donation amount is :",data2["Donation Amount"].min())

# Get mean donation amount
print("Mean donation amount is: ", data2["Donation Amount"].mean())

# Get median donation amount
print("Median donation amount is: ", data2["Donation Amount"].median())

# Get seventy fith percentile donation amount
seventy_fifth = donations["Donation Amount"].quantile(0.75)
print("75th percentile is: ", seventy_fifth)

# Get twenty fith percentile donation amount
twenty_fifth = donations["Donation Amount"].quantile(0.25)
print("25th percentile is: ", twenty_fifth)

# Identifying thresholds
donation_iqr = seventy_fifth - twenty_fifth
print("IQR is: ",donation_iqr)

upper = seventy_fifth + (1.5 * donation_iqr)
print("Upper limit is: ",upper)

lower = twenty_fifth - (1.5 * donation_iqr)
print("Lower limit is: ",lower)
```

```python
# Exclude outliers from data2 dataframe
data2_no_outliers = data2[(data2["Donation Amount"] > lower) & (data2["Donation Amount"] < upper)][["Project ID","School ID","Donation Amount"]]
print("Minimum donation amount is :",data2_no_outliers["Donation Amount"].min())
print("Maximum donation amount is :",data2_no_outliers["Donation Amount"].max())
print(data2_no_outliers.shape)

data2_no_outliers.head()
```

```python
# Get outliers from data2 dataframe
data2_outliers = data2[(data2["Donation Amount"] < lower) | (data2["Donation Amount"] > upper)][["Project ID","School ID","Donation Amount"]]
print("Minimum donation amount is :",data2_outliers["Donation Amount"].min())
print("Maximum donation amount is :",data2_outliers["Donation Amount"].max())
print(data2_outliers.shape)

data2_outliers.head()
```

```python
# In which states there are more donations done by donors. 
data3 = pd.merge(donations, donors, how="inner", on="Donor ID")
data3[["Donor State","Donor ID","Donation Amount"]]
s2 = data3.groupby("Donor State")["Donation ID"].agg("count").sort_values(ascending=False).head(10)
print(s2)
```

```python
# Is there a relationship between the number of projects offered and number of donations made by the donors. 
# Which states performing better in this case ? 
# How many of them responding project requests below average and which states are performing best in terms of donations per project ? 
#In order to answer this question we must first get the number of projects per state and then number of donations made per state. 
# Then we should merge this two and plot a scatter plot to visualize it.

nu_of_projects = schools["School State"].value_counts()
nu_of_donations_by_donor = data3.groupby("Donor State")["Donation ID"].count().sort_values(ascending=False)

df = pd.concat([nu_of_projects,nu_of_donations_by_donor], axis = 1, keys = ["Projects", "Donations"])
df.head(10)
```

```python
# Returns column with any Nan values 
df.loc[:,df.isnull().any()]
df = df.dropna()
```

```python
# Now it is time to visualize this data for further insights.
df.plot.scatter(x = "Projects", y = "Donations")
plt.title("Projects vs Donations")
for i, j in enumerate(df.index):
    plt.annotate(j,(df.Projects[i], df.Donations[i]))

slope, intercept = np.polyfit(df.Projects, df.Donations, 1)
x = np.array([df.Projects.min(), df.Projects.max()])
y =  slope * x + intercept 
_ = plt.plot(x,y)
_ = plt.tight_layout()
_ = plt.margins(0.05)
# Increase figure size : 
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
```

```python
# What is the mean time that takes a project to be fully funded after posted and how it varies between states ?

data5 = pd.merge(projects, schools, how="inner", on="School ID")

# How many projects are still open, we may choose to remove that ones
print(data5[["Project Posted Date", "Project Fully Funded Date"]].isnull().sum())

# Convert column from object to datatype
data5["Project Fully Funded Date"] = pd.to_datetime(data5["Project Fully Funded Date"]).dropna()
data5["Project Posted Date"] = pd.to_datetime(data5["Project Posted Date"])

# Add "Funding Time"
data5["Funding Time"] = data5["Project Fully Funded Date"] - data5["Project Posted Date"]

# Consult Dates
print("\n", data5[["Funding Time","Project Fully Funded Date","Project Posted Date"]].head())

# Check for Null Dates
print("\n", data5[["Funding Time","Project Fully Funded Date","Project Posted Date"]].isnull().sum())

# Drop NaT values.
data6 = data5[pd.notnull(data5["Funding Time"])]

# Check for Null Dates
print("\n", data6[["Funding Time","Project Fully Funded Date","Project Posted Date"]].isnull().sum())

data6["Funding Time"] = data6["Funding Time"].dt.days

# Consult Dates
print("\n", data6[["Funding Time","Project Fully Funded Date","Project Posted Date"]].head())
```

```python
# Consult Dates
print(data6[["Funding Time","Project Fully Funded Date","Project Posted Date"]].head())

# Mean funding time per Project ID
mean_time_project_funding = data6.groupby("Project ID")["Funding Time"].mean()
print("\n", mean_time_project_funding)

# Overall Project ID
overall_mean_time = mean_time_project_funding.mean()
print("\n", overall_mean_time)
```

```python
# Mean funding time per Project ID by State
states_project_funding_time = data6.groupby(["School State", "Project ID"])["Funding Time"].mean()
print(states_project_funding_time)
```

```python
# In which 10 states project funded earlier than average days ?
states_average_funding_time = states_project_funding_time.groupby("School State").mean()
ss = states_average_funding_time.round(0)
ss[ss < 32].sort_values().head(10)
```

```python
# Plot the dataframe

fast_funding_states = ss[ss < 32].sort_values().head(10)
fast_funding_states.plot.bar()
plt.axhline(32, color ="m", linewidth = 2)
plt.ylim(0,40)
plt.xlabel("States")
plt.ylabel("Fully Funding Time  ( in days )")
plt.title("States that fund projects faster than others")
```
